# Random Numbers

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

### One Random Number:

```python
np.random.random()
```

### Several Random Numbers:

```python
np.random.random_sample(10)
```

```python
y = np.random.random_sample(100)
plt.plot(y, 'o')
plt.show()
```

```python
y = np.random.random_sample(1000000)
n, bins, patches = plt.hist(y, 100)
plt.show()
```

### Adding Random Numbers Together:

```python
NumPoints = 100
z = np.random.random_sample(NumPoints) + np.random.random_sample(NumPoints)
plt.plot(z, 'o')
plt.show()
```

```python
NumPoints = 1000000

NumLoops = 7
z = np.zeros(NumPoints)
for i in range(NumLoops):
    z = z + np.random.random_sample(NumPoints)
n, bins, patches = plt.hist(z, 100)
plt.show()
```
